## Removing Outliers with Z-Score Method

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df1))
z
```

```python
np.where(z>3)
```

```python
ad=df1[(z<3).all(axis=1)]
ad
```

```python
print('The data loss percentage after removing outliers using zscore method ',((df1.shape[0]-ad.shape[0])/df1.shape[0])*100)
```

## Removing Outliers using IQR method

```python
Q1=df1.quantile(0.25)
Q3=df1.quantile(0.75)
IQR=Q3-Q1
df2=df1[~((df1<(Q1-1.5*IQR))|(df1>(Q3+1.5*IQR))).any(axis=1)]
```

```python
print('The data loss percentage after removing outliers using IQR method ',((df1.shape[0]-df2.shape[0])/df1.shape[0])*100)
```

Z Score method has less data loss

```python
model = RandomForestClassifier()
param_grid = {
    'n_estimators': [200,500],
    'max_features': [.5,.7],
    'bootstrap': [False, True],
    'max_depth':[3,6],
    'criterion':['gini','entropy']
}
grid_search = GridSearchCV(model, param_grid, n_jobs=-1, cv=2)
grid_search.fit(a_train,b_train)
```
